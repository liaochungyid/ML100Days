- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans

estimators = [('k_means_8', KMeans(n_clusters=6)),
              ('k_means_3', KMeans(n_clusters=3)),
              ('k_means_1_init', KMeans(n_clusters=3, n_init=1,init='random')),
              ('k_means_8_init', KMeans(n_clusters=3, n_init=8,init='random')),
              ('k_means_++_init', KMeans(n_clusters=3,init='k-means++'))]
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D

fignum=1
titles = ['8 clusters', '3 clusters', '3 clusters 1 init', '3 clusters 8 init', '3 clusters ++ init']
for name, est in estimators:
    fig = plt.figure(fignum,figsize=(4,3))
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
    
    est.fit(X)
    labels = est.labels_
    ax.scatter(X[:,0],X[:,1],X[:,2],c=labels.astype(np.float), edgecolor='k')
    
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum += 1
    
```
